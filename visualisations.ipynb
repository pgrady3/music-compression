```python
import numpy as np
import matplotlib.pyplot as plt

from utils.io import read_mp3
```

## Loading the data

```python
model_inp_path = 'data/sample/fma_xs/model_audio/type1_loss/sample_input.mp3'
model_out_path = 'data/sample/fma_xs/model_audio/type1_loss/sample_output.mp3'
```

```python
_, audio_inp = read_mp3(model_inp_path)
_, audio_out = read_mp3(model_out_path)
```

```python
print('Input mean = {}'.format(np.mean(audio_inp)))
print('Output mean = {}'.format(np.mean(audio_out)))

print('Input std = {}'.format(np.std(audio_inp)))
print('Output std = {}'.format(np.std(audio_out)))
```

```python
norm_out = audio_out - np.mean(audio_out)
norm_out = norm_out/np.std(norm_out)

mean_in = np.mean(audio_inp)
audio_out_scaled = norm_out * np.std(audio_inp - mean_in)
audio_out_scaled = audio_out_scaled + mean_in
```

```python
# plot in time and frequency domain
fig, ax = plt.subplots(2, 2, figsize=(15, 15))

ax[0][0].plot(audio_inp)
ax[0][0].set_title('Input [time domain]')

ax[0][1].specgram(audio_inp, NFFT=512, noverlap=256, Fs = 40e3)
ax[0][1].set_title('Input [frequency domain]')

ax[1][0].plot(audio_out_scaled)
ax[1][0].set_title('Ouput [time domain]')

ax[1][1].specgram(audio_out_scaled, NFFT=512, noverlap=256, Fs = 40e3)
ax[1][1].set_title('Output [freq domain]')

plt.show()
```
