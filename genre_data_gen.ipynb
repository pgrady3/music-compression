```python
import pandas as pd
import numpy as np
import glob
import os

import csv
```

```python
track_file = 'data/sample/fma_xs_metadata/tracks.csv'
genre_file = 'data/sample/fma_xs_metadata/genres.csv'
```

```python
tracks = pd.read_csv(track_file, index_col=0, header=[0, 1], keep_default_na=False)
genre = pd.read_csv(genre_file, index_col=0, header=[0], keep_default_na=False)
```

```python
genre_top_df = tracks['track']['genre_top']
```

```python
genre_top_df
```

```python
genre_top_df.replace('', np.nan, inplace=True)
genre_top_df.dropna(inplace=True)
```

```python
genre_top_df.value_counts()
```

```python
unique_genres = genre_top_df.unique()
genre_labels_map = dict(zip(unique_genres, range(len(unique_genres))))
```

```python
print(genre_labels_map)
```

```python
# read all the file names of .mp3
file_paths = glob.glob('data/sample/fma_xs/test/*.mp3') + glob.glob('data/sample/fma_xs/train/*.mp3')
file_names = [int(os.path.basename(f).split('.')[0]) for f in file_paths]
```

```python
genre_dict = genre_top_df.to_dict()
```

```python
subset_dict = {file_name: genre_labels_map[genre_dict[file_name]] for file_name in file_names}
```

```python
print(subset_dict)
```

```python
with open('file_to_label.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in subset_dict.items():
       writer.writerow([key, value])
```

```python
with open('genre_to_label.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in genre_labels_map.items():
       writer.writerow([key, value])
```
